## Problema 1:

Desarrolle una función la cual reciba dos parámetros, una imagen y un entero llamado color, la función debe devolver una imagen la cual tenga activos los canales de color según los siguientes puntos:

Si el parámetro color vale 1, la imagen debe mostrar activos únicamente el color azul.
Si el parámetro color vale 2, la imagen debe mostrar activos únicamente el color verde.
Si el parámetro color vale 3, la imagen debe mostrar activos únicamente el color rojo.
Si el parámetro color vale 10, la imagen debe mostrar activos únicamente los colores rojo y verde.
Si el parámetro color vale 20, la imagen debe mostrar activos únicamente los colores verde y azul.
Si el parámetro color vale 30, la imagen debe mostrar activos únicamente los colores azul y rojo.

```python
import matplotlib.pyplot as plt
import cv2

def display_image_color(image, color):
    # Cargar la imagen
    img = cv2.imread(image)

    if img is None:
        print("Error al cargar la imagen.")
        return None

    # Crear una copia de la imagen
    selected_color_image = img.copy()

    # Definir los canales de color que deben estar activos para cada valor de color
    if color == 1: # Azul
        selected_color_image[:, :, 1] = 0 # Verde
        selected_color_image[:, :, 2] = 0 # Rojo
    elif color == 2: # Verde
        selected_color_image[:, :, 0] = 0 # Azul
        selected_color_image[:, :, 2] = 0 # Rojo
    elif color == 3: # Rojo
        selected_color_image[:, :, 0] = 0 # Azul
        selected_color_image[:, :, 1] = 0 # Verde
    elif color == 10: # Rojo y verde
        selected_color_image[:, :, 0] = 0 # Azul
    elif color == 20: # Verde y azul
        selected_color_image[:, :, 2] = 0 # Rojo
    elif color == 30: # Azul y rojo
        selected_color_image[:, :, 1] = 0 # Verde
    else:
        print("Valor de color no válido.")
        return None

    # Convertir la imagen de BGR a RGB
    selected_color_image = cv2.cvtColor(selected_color_image, cv2.COLOR_BGR2RGB)

    # Mostrar la imagen resultante
    plt.imshow(selected_color_image)
    plt.show()

# Prueba la función con la imagen de 'bmw.jpg' y el color 1
display_image_color('/content/sample_data/bmw.jpg', 30)
```

## Problema 3:

Cree una función que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su función debe crear 3 imágenes como salida. Para entregar este ejercicio debe incluir una las imágenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema #2.

```python
# Cargar la imagen
image = Image.open('/content/sample_data/bmw.jpg')

# Dividir la imagen en canales de color
r, g, b = image.split()

# Convertir cada canal de color a escala de grises
r = r.convert("L")
g = g.convert("L")
b = b.convert("L")

# Guardar las imágenes en escala de grises
r.save("bmw_gray_red.jpg")
g.save("bmw_gray_green.jpg")
b.save("bmw_gray_blue.jpg")

# Mostrar las imágenes en escala de grises
plt.figure(figsize=(12, 12))

plt.subplot(2, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(r, cmap='gray')
plt.title('Red Channel in Gray Scale')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(g, cmap='gray')
plt.title('Green Channel in Gray Scale')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel in Gray Scale')
plt.axis('off')

plt.show()
```

## Problema 4:

Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmético para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media de la distribución.

```python
def show_histograms(image_path):
    # Cargar la imagen
    image = Image.open(image_path)

    # Dividir la imagen en canales de color
    r, g, b = image.split()

    # Convertir la imagen a escala de grises usando un promedio aritmético de los canales de color
    gray = Image.fromarray((np.array(r) + np.array(g) + np.array(b)) / 3).convert("L")

    # Crear una figura para los histogramas
    plt.figure(figsize=(15, 10))

    for i, (channel, color) in enumerate(zip([r, g, b, gray], ['red', 'green', 'blue', 'gray']), start=1):
        # Convertir el canal de color a una matriz numpy
        channel_array = np.array(channel)

        # Calcular el histograma y la media
        hist, bins = np.histogram(channel_array.flatten(), bins=256, range=[0,256])
        mean = channel_array.mean()

        # Crear un subplot para el histograma del canal de color
        plt.subplot(2, 2, i)
        plt.fill_between(bins[:-1], hist, color=color, alpha=0.7)
        plt.axvline(mean, color='k', linestyle='dashed', linewidth=1)
        plt.title(f'{color.capitalize()} Histogram')
        plt.xlabel('Pixel Value')
        plt.ylabel('Frequency')
        plt.tight_layout()

# Ejecutar la función para la imagen del bmw
show_histograms('/content/sample_data/bmw.jpg')
```
